```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import sklearn.datasets

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the data from sklearn
cancer_data= sklearn.datasets.load_breast_cancer()
cancer_data
```

```python
# loading the data to a data frame
#data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
data=pd.DataFrame(cancer_data.data,columns=cancer_data.feature_names)
data
```

```python
# adding the 'target' column to the data frame
data['label'] = cancer_data.target
```

```python
data.info()
```

```python
print('rows: ',data.shape[0])
print('columns: ',data.shape[1])
```

```python
data.isnull().sum()
```

```python
# statistical measures about the data
data.describe().T
```

```python
# checking the distribution of Target Varibale
data['label'].value_counts()
```

```python
# 1-> Bengin 
# 0-> Malignant 
```

```python
data.groupby('label').mean()
```

```python
# Separating the features and target
x=data.drop(columns='label',axis=1)
y=data['label']
```

```python
x
```

```python
y
```

```python
# Splitting the data into training data & Testing data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
# Model Training 
```

```python
# Logestic regression 
```

```python
lr= LogisticRegression()

lr.fit(x_train,y_train)
```

```python
# Model Evaluation
```

```python
# Accuracy Score on train data
x_train_pred=lr.predict(x_train)
train_acc=accuracy_score(y_train,x_train_pred)
print('accuracy_Score: ',train_acc)
```

```python
# Accuracy Score on test data
x_test_pred=lr.predict(x_test)
test_acc=accuracy_score(y_test,x_test_pred)
print('accuracy_score: ',test_acc)
```
